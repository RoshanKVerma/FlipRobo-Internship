7. Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location


Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
# Creating an instance of webdriver for google chrome
driver=webdriver.Chrome(r"C:\Users\RoshanV\Desktop\Internship\Assignment 2\chromedriver.exe")
driver.maximize_window()

# Using webdriver we'll now open the website in chrome
driver.get("https://www.naukri.com/naukri-dot-com-recruiters")

# Find using xpath
designation=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[2]/div/form/div[1]/div/div[1]/div[1]/div[2]/input")
designation.send_keys('Data Science') # Typing Data Science

# Find using xpath
search=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/div/form/div[1]/button')
search.click()


# Making empty list for our data to be dumped
JobDesignation=[]
JobLoc=[]
CompanyName=[]
skills=[]
Name=[]

time.sleep(10)

# Extracting Data from The Website using XPATH and Dumping it in respective Lists

jtt=driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
for i in jtt:
    title=i.text
    JobDesignation.append(title)
    
jlt=driver.find_elements(By.XPATH,'//small[@class="ellipsis"]')
for i in jlt:
    location=i.text
    JobLoc.append(location)
    
cnt=driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
for i in cnt:
    skill=i.text
    skills.append(skill)
    
cnt=driver.find_elements(By.XPATH,'//a[@class="ellipsis"]')
for i in cnt[1::2]:
    compname=i.text
    CompanyName.append(compname)
    
ret=driver.find_elements(By.XPATH,'//span[@class="fl ellipsis"]')
for i in ret:
    exp=i.text
    Name.append(exp)
    


# Using Pandas to Create DataFrame of the various Lists

df=pd.DataFrame({'Name':Name,'Designation':JobDesignation,'Job_Location':JobLoc,'Skills':skills,'Company_Name':CompanyName})

df
```
